```python
# Extract events from the first page
events = data["_embedded"]['events']
csv_data.extend(events)

# Check if there are more pages of events
#while 'next' in data['page']['number']:
 #   next_page_url = data['pagination']['next']
  #  response = requests.get(next_page_url)
   # data = response.json()
    
    # Add new events from the next page
   # events = data['_embedded']['events']
   # csv_data.extend(events)

   # print(f"Retrieved {len(events)} more events, total: {len(all_events)} events.")

# Now all_events contains all retrieved events
#print(f"Total events retrieved: {len(all_events)}")

# Define CSV headers
csv_headers = ['Event ID', 'Event Name', 'Start Date/Time', 'End Date/Time', 'Event URL', 'Status', 'Genre', 'Venue', 'Address', 'City', 'State', 'Postal Code', 'Longitude', 'Latitude']
csv_data.append(csv_headers)

# Set to track seen event IDs to avoid duplicates
seen_event_ids = set()

# Loop through the events
for event in events:
    event_id = event['id']
    
    # Skip events that have already been added
    if event_id not in seen_event_ids:
        continue
    
    # Add the event ID to the seen set
    seen_event_ids.add(event_id)
                      
    event_id = event['id']
    event_name = event['name']
    start_date = event['sales']['public']['startDateTime']  # The event's local start time
    end_date = event['sales']['public']['endDateTime']
    event_url = event['url']
    status = event['dates']['status']['code']
    genre = event['classifications'][0]['segment']['name']
    venue = event['_embedded']['venues'][0]['name']
    address = event['_embedded']['venues'][0]['address']['line1']
    city = event['_embedded']['venues'][0]['city']['name']
    state = event['_embedded']['venues'][0]['state']['name']
    postal_code =event['_embedded']['venues'][0]['postalCode']
    longitude = event['_embedded']['venues'][0]['location']['longitude']
    latitude = event['_embedded']['venues'][0]['location']['latitude']
    
    # Append a row for each event
    csv_data.append([event_id, event_name, start_date, end_date, event_url, status, genre, venue, address, city, state, postal_code, longitude, latitude])

# Step 3: Write the data to a CSV file
csv_filename = 'events_data.csv'
with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(csv_data)

print(f"CSV file '{csv_filename}' has been written successfully.")
```
